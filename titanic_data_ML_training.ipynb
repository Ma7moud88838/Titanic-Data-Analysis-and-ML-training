```python
df.drop(["PassengerId", "Name", "Cabin", "Ticket"], axis = 1, inplace=True)
```

```python
df.describe()
```

```python
df['Age'].fillna(df["Age"].mean(), inplace=True)
```

```python
df.dropna(inplace = True)
df.drop_duplicates(inplace=True)
df.isnull().sum()
```

```python
df.info()
```

```python
data = pd.get_dummies(df, columns=["Sex", "Embarked"], dtype=np.int64)
data.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
cor = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(cor, annot=True, cmap='coolwarm', center=0)
```

```python
data.drop(["Sex_male", "Embarked_S"], axis=1, inplace=True)
data
```

```python
col = list(data.columns)
col.remove("Survived")
X_tot = np.array(data[col])
y_tot = np.array(data["Survived"])
print(X_tot)
print(X_tot.shape)
print(y_tot)
print(y_tot.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tot, y_tot, shuffle=True, test_size=0.2, random_state=42)
print(X_train.shape)
print(y_train.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```

```python
from sklearn.svm import SVC
model = SVC(kernel="linear")
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(7)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```
